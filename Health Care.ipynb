# Linear Regression

```python
#Lets take the scaled dataset predict the target "Diabetes" by building a linear model
df_Scaled.head()
```

```python
# Lets split the Independent and dependent variables
df_x=df_Scaled.drop('BMI 1Y', axis=1)
df_x.head()
```

```python
df_y=df_Scaled.iloc[:,5]
df_y
```

Splitting the data for training and testing phases.
Since my dataset is very small. I splitted the 80% of data for training and 20% for testing.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.20, random_state=0)
```

```python
# Importing Linear Regression Model
from sklearn.linear_model import LinearRegression
LR=LinearRegression()

# Training my model
LR.fit(X_train,y_train)
y_pred=LR.predict(X_test)
print("Regression Score:", LR.score(X_test,y_test))
```

```python
from sklearn.metrics import mean_squared_error,r2_score
print("MSE is", mean_squared_error(y_pred,y_test))
print("R-Square is", r2_score(y_test,y_pred))
```

The model is able to predict the BMI of a patient after an year of surgery with 49% accuracy.
